##**3.ANALYSIS BASED ON MOTIVE OF ATTACK**

---

Behind every any volurnaly action a specific reason is there. So let's find the reason behind the terrorist attacks.

```python
#considering motive column.
motive_df = pd.DataFrame(df_new,columns = ["motive"])
motive_df.dropna(inplace=True)
motive_df
```

```python
from wordcloud import WordCloud, STOPWORDS
comment_words = ''
stopwords = set(STOPWORDS)
for val in motive_df.motive:
     
    # typecaste each val to string
    val = str(val)
 
    # split the value
    tokens = val.split()
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "

wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
 
# plot the WordCloud image                      
plt.figure(figsize = (7,7), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
